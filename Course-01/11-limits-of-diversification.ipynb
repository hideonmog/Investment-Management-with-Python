# Limits of Diversificaton

We know the advanatges of diversification, it allows us to eliminate the unrewarded idiosyncratic risks in our portfolios; it's okay to hold risk as long as it is rewarded. However, this works on average across market conditions, diversification will fail when you most need it. If we are in a severe bear market (e.g 2008 crash) and all stocks are going down, diversifiying our portfolio won't do much for us. One cannot diversify away systematic risk.

### Diversification vs Hedging 

Hedging (i.e avoiding risk taking) is the only effective way to obtain downside protection. The problem with hedging is that investors give up on the upside at the same time as they give up on the downside. Hedging is okay, but more so for wealthy investors.

### Hedging vs Insurance

Insurance allows for downside protection whilst allowing for upside potential. Insurance is dynamic hedging, it gives you the best of both worlds. You get downsid protection but can still enjoy a fair amount of upside potential.

### Rolling Correlations - along with MultiIndexes and `.groupby`

```python
ts_corr = ind_return.rolling(window = 36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
ts_corr.index.names = ['date', 'industry']
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level = 'date').apply(lambda cormat: cormat.values.mean())
ind_tr36corr.plot()
```

```python
tmi_tr36mrets.plot(figsize = (15, 8), label = 'Tr 36 mo rets', legend = True)
ind_tr36corr.plot(label = 'Tr 36 mo corr', legend = True, secondary_y = True)
```
