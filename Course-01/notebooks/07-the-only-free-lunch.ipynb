# The Only Free Luch in Finance

## 2-Asset Combinations

Say we have two assets A and B with a certain return and volatility associated with them. If we have a portfolio with 100% in A, the return and volatility of the portfolio would just be that of A's, and the same can be done with B. The problem lies when we have differnt percentages in different assets. The returns would just be the weighted average of the returns of A and B, but the volatility would be totally different, unless A and B are perfectly correlated. 

For a portfolio with assets A and B the return can be calculated as

$$R\left(w_A, w_B \right) = w_A \times R_A + w_B \times R_B,$$

where $w$ is the weighted average and $R$ is the return of the associated asset. 

The expression for the volatility is a lot more complicated and is give by the following equation

$$\sigma^2\left(w_A, w_B\right) = \sigma^2_Aw^2_A + \sigma^2_Bw^2_B +2w_Aw_B\sigma_A\sigma_B\rho_{A,B},$$

where $\sigma^2$ is the variance of the associated asset and $\rho$ is the measure of correlation between assets.

Correlation between two assets can usually be represented as a number between 0 and 1, if we model a portfolio of two assets for different correlations we would see that at some correlation level the portfolio volatility is actually lower than both of the assets' individual volatilities. This is what's known as 'the only free lunch in finance'. We are able to create a lower risk portfolio from two higher risk assets due to them being de-correlated. The lowest possible volatility you can construct from a portfolio of two assets A and B increases as the correlation between A and B increases. This is the power of diversification.
